```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
my = pd.read_csv('maoyanfilms_clean.csv')
```

```python
fr1 = my[['films_id','films_box_office']].dropna()
fr1
```

```python
fr2 = my[['films_id','films_score']].dropna()
fr2
```

```python
fr3 = pd.merge(fr1,fr2,on='films_id')
```

```python
fr3
```

```python

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#绘制图片
plt.scatter(fr3['films_score'],fr3['films_box_office'])

#添加描述信息
plt.xlabel("评分",fontproperties=my_font)
plt.ylabel("票房",fontproperties=my_font)
plt.title("票房与评分的关系",fontproperties=my_font)

#保存图片
plt.savefig("./票房与评分的关系.png")

#展示图片
plt.show()
```
