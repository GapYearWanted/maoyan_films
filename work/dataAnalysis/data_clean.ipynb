```python
import pandas as pd
import numpy as np
import re
my = pd.read_csv('maoyanfilms.csv')
```

# 数据清洗

```python
 my
```

### 首映时间的处理

```python
# 去掉后面的中文
def handle_film_first_time(x):
    pattern = re.compile(r'[\u4e00-\u9fa5]')
    date = re.sub(pattern,"",str(x))
    return date
my['film_first_time'] = my['film_first_time'].apply(handle_film_first_time)
```

### 评分、评论数、票房 异常数据处理


```python
# 评分、评论数 中含有异常数据 “. 2.”等
score_list = []
for i in range(0,10):
    for  j in range(0,10):
        num = str(i)+"."+str(j)
        score_list.append(num)
series_a = my[~my['films_score'].isin(score_list)]['films_score']
series_b = series_a.dropna()
list_a = series_b.index
for index in list_a:
    my.loc[index,'films_score']= np.nan
    my.loc[index,'films_comment_num'] = np.nan
    my.loc[index,'films_box_office'] = '暂无'
```

### 演员、导演名字处理

```python
my
```

```python
# 去掉演员、导演名字之间的空格
def remove_space(x):
    s = str(x).replace(' ', '')
    return s
my['films_actor'] = my['films_actor'].apply(remove_space)
my['films_director'] = my['films_director'].apply(remove_space)
```

```python
my['films_actor'] = my['films_actor'].apply(lambda x : str(x).replace(' ','').replace('\n',''))
my['films_director'] = my['films_director'].apply(lambda x : str(x).replace(' ','').replace('\n',''))
```

```python
 my
```

### 票房处理 （单位万）

```python
def handle_films_box_office	(x):
    s = x[-1:]
    if s =='亿':
        return int(float(x[0:len(x)-1])*10000)
    elif s =='万':
        try:
            f = float(x[0:len(x)-1])
        except:
            return np.nan
        else:          
            return f
    else:
        return np.nan
my['films_box_office']= my['films_box_office'].apply(handle_films_box_office)
```
